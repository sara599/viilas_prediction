```python
import csv
import sqlite3
```

```python
# Import required modules


# Connecting to the geeks database
connection = sqlite3.connect('data_db11.db')

# Creating a cursor object to execute
# SQL queries on a database table
cursor = connection.cursor() #make excution

# Table Definition
create_table = '''CREATE TABLE villa(
                price REAL NOT NULL,
                location TEXT NOT NULL,
                title TEXT NOT NULL,
                link TEXT NOT NULL,
                viwers INTEGER NOT NULL,
                negotiable INTEGER NOT NULL,
                Bedrooms INTEGER NOT NULL,
                Bathrooms INTEGER NOT NULL,
                Area REAL NOT NULL,
                Type TEXT NOT NULL,
                Compound TEXT NOT NULL,
                Delivery_Term TEXT NOT NULL,
                finished TEXT NOT NULL,
               Delivery_Date TEXT NOT NULL,
               Down_Payment REAL NOT NULL,
                Day INTEGER NOT NULL,
                Month INTEGER NOT NULL,
                Year INTEGER NOT NULL,
                Season TEXT NOT NULL,
                country TEXT NOT NULL,
                Cash REAL NOT NULL,
                Installment REAL NOT NULL,
                Balcony INTEGER NOT NULL,
                Pets_Allowed INTEGER NOT NULL,
                Private_Garden INTEGER NOT NULL,
                Security INTEGER NOT NULL,
                Electricity_Meter INTEGER NOT NULL,
                Water_Meter INTEGER NOT NULL,
                Natural_Gas INTEGER NOT NULL,
                Landline INTEGER NOT NULL,
                Built_in_Kitchen INTEGER NOT NULL,
                Appliances INTEGER NOT NULL,
                Maids_Room INTEGER NOT NULL,
                Covered_Parking INTEGER NOT NULL,
                Central_conditioner_heating INTEGER NOT NULL,
                Pool INTEGER NOT NULL,
                Elevator INTEGER NOT NULL
                
                );'''



# Creating the table into our
# database
cursor.execute(create_table)

# Opening the vilas-records.csv file
file = open('villas_after_editing.csv',encoding="utf8")

# Reading the contents of the
# person-records.csv file
contents = csv.reader(file)

# SQL query to insert data into the
# villa table
insert_records = "INSERT INTO villa(price,location,title,link,viwers,negotiable,Bedrooms,Bathrooms,Area,Type,Compound,Delivery_Term,finished,Delivery_Date,Down_Payment,Day,Month,Year,Season,country,Cash,Installment,Balcony, Pets_Allowed,Private_Garden,Security,Electricity_Meter,Water_Meter,Natural_Gas, Landline,Built_in_Kitchen,Appliances,Maids_Room,Covered_Parking,Central_conditioner_heating,Pool,Elevator ) VALUES(?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)"

# Importing the contents of the file
# into our villa table
cursor.executemany(insert_records, contents)
# SQL query to retrieve all data from
# the person table To verify that the
# data of the csv file has been successfully
# inserted into the table
select_all = "SELECT * FROM villa"
rows = cursor.execute(select_all).fetchall()

# Output to the console screen
for r in rows:
    print(r)

# Committing the changes
connection.commit()

# closing the database connection
connection.close()
```
